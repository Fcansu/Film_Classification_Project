```python
from moviepy.video.io.VideoFileClip import VideoFileClip

# Video dosyasının yolu
video_path = "/content/drive/MyDrive/Colab Notebooks/sample_data/ACI KAHVE - FRAGMAN.mp4"

# Video dosyasını yükle
clip = VideoFileClip(video_path)

# Toplam süre (saniye cinsinden)
total_duration = clip.duration

# Aralık (120 saniye = 2 dakika)
interval = 2

# Kaydedilecek görüntülerin klasörü
output_folder = "screenshots"
import os
os.makedirs(output_folder, exist_ok=True)

# 2 dakika aralıklarla görüntü al ve kaydet
for t in range(0, int(total_duration), interval):
    frame = clip.get_frame(t)
    output_path = os.path.join(output_folder, f"frame_{t//60}min_{t%60}sec.jpg")
    from PIL import Image
    Image.fromarray(frame).save(output_path)

print(f"2 sn aralıklarla görüntüler {output_folder} klasörüne kaydedildi.")
```

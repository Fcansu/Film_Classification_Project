```python
import pandas as pd
```

```python
df = pd.read_csv("data_siniflanmis.csv")
df.head()
```

```python
df['degisken 1'].fillna('', inplace=True)
df['degisken 2'].fillna('', inplace=True)
df['degisken 3'].fillna('', inplace=True)
df['degisken 4'].fillna('', inplace=True)
```

```python
df.head(10)
```

```python
df['film_kategori'] = df.apply(lambda x: [x['degisken 1'], x['degisken 2'], x['degisken 3'], x['degisken 4']], axis=1)
df.head()
```

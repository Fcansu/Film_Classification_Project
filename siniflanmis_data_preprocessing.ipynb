```python
import pandas as pd
```

```python
df = pd.read_csv("data_siniflanmis.csv")
df.head()
```

```python
df['degisken_1'].fillna('', inplace=True)
df['degisken_2'].fillna('', inplace=True)
df['degisken_3'].fillna('', inplace=True)
df['degisken_4'].fillna('', inplace=True)
```

```python
df.head(10)
```

```python
df['film_kategori'] = df.apply(lambda x: [x['degisken_1'], x['degisken_2'], x['degisken_3'], x['degisken_4']], axis=1)
df.head()
```
